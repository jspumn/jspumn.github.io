```python
# üìå Step 1: ÂÆâË£ùÂøÖË¶ÅÂ•ó‰ª∂ÔºàÂä†‰∏ä lxml_html_cleanÔºâ
!pip install newspaper3k
!pip install sentence-transformers
!pip install lxml_html_clean


# üìå Step 2: ÂåØÂÖ•ÂáΩÂºèÂ∫´
from newspaper import Article
from sentence_transformers import SentenceTransformer, util

# üìå Step 3: ÂàùÂßãÂåñ embedding Ê®°ÂûãÔºàÂèØÈÅ∏ÊèõÂÖ∂‰ªñÊ®°ÂûãÔºâ
model = SentenceTransformer('sentence-transformers/all-MiniLM-L6-v2')

# üìå Step 4: Êì∑ÂèñÊñ∞ËÅûÂÖßÂÆπÂáΩÂºè
def get_article_text(url):
    try:
        article = Article(url)
        article.download()
        article.parse()
        return article.text
    except Exception as e:
        return f"Error fetching article: {e}"

# üìå Step 5: Ëº∏ÂÖ•ÂÖ©ÂÄãÊñ∞ËÅûÈÄ£Áµê
url1 = input("Enter the first news URL: ")
url2 = input("Enter the second news URL: ")

text1 = get_article_text(url1)
text2 = get_article_text(url2)

print("\n--- News 1 Preview ---\n", text1[:500])
print("\n--- News 2 Preview ---\n", text2[:500])



##test1 : url1: https://avinews.com/ph/new-vaccine-development-center-inaugurated-in-the-philippines/
##test2 : url2: https://tribune.net.ph/2025/04/15/vaccine-center-combats-animal-diseases
##test3 : url3: https://www.bworldonline.com/economy/2025/04/13/665630/meat-imports-top-16-in-feb-as-asf-continues-to-pressure-supply/#google_vignette

# üìå Step 6: ÂêëÈáèÂåñ‰∏¶Ë®àÁÆóË™ûÊÑèÁõ∏‰ººÂ∫¶
embedding1 = model.encode(text1, convert_to_tensor=True)
embedding2 = model.encode(text2, convert_to_tensor=True)
similarity = util.cos_sim(embedding1, embedding2).item()

# üìå Step 7: ÁµêÊûúËº∏Âá∫
print(f"\n‚úÖ Cosine similarity: {similarity:.3f}")

if similarity > 0.90:
    print("‚úÖ SAME EVENT ‚Äì High Confidence")
elif similarity > 0.75:
    print("‚ö†Ô∏è POSSIBLY Related ‚Äì Needs Review")
elif similarity > 0.50:
    print("‚ùî Possibly Similar Topic ‚Äì Likely Different Event")
else:
    print("‚ùå DIFFERENT Event")

```
